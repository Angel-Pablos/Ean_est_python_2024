## Cul es el objetivo de la inferencia?

## Tipos de muestreo

### Muestreo Aleatorio Simple

El muestreo aleatorio simple es el mtodo ms bsico y comn de muestreo. En este mtodo, cada miembro de la poblacin tiene la misma **probabilidad de ser seleccionado en la muestra**. Esto significa que cada muestra de tamao $n$ tiene la misma probabilidad de ser seleccionada de la poblacin.



```python
import random

# Poblacin con 10 elementos
poblacion = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Tamao de la muestra
tam_muestra = 5

# Muestreo aleatorio simple
muestra = random.sample(poblacion, tam_muestra)

print("Muestra:", muestra)
```

### Muestreo Aleatorio Simple Sin Reemplazo

En este mtodo, cada elemento seleccionado para la muestra es eliminado de la poblacin para futuras selecciones. Esto significa que no puede ser vuelto a seleccionar, lo que reduce el tamao de la poblacin de donde se eligen los sucesivos elementos.

**Ventajas:**

* Cada miembro de la poblacin tiene la misma probabilidad de ser elegido.
* Los resultados son muestras dependientes, ya que la seleccin de un miembro afecta la seleccin del siguiente.

**Implicaciones Estadsticas:**

La covarianza entre indicadores de seleccin para diferentes elementos de la muestra es negativa, pues la seleccin de un elemento disminuye la probabilidad de seleccionar otros.

**Cundo usar?**

Este enfoque es muy comn cuando el tamao de la poblacin es conocido y es relativamente pequeo.


### Muestreo Aleatorio Simple Con Reemplazo

En el muestreo con reemplazo, cada elemento seleccionado para la muestra es "reemplazado" en la poblacin antes de realizar la siguiente seleccin, lo que significa que puede ser seleccionado ms de una vez.

**Ventajas:**

* Simplifica el anlisis estadstico porque todas las muestras son independientes entre s.


**Implicaciones Estadsticas:**

La covarianza entre indicadores de seleccin para diferentes elementos de la muestra es cero, ya que la seleccin de un elemento no afecta la probabilidad de seleccionar ese mismo elemento u otros nuevamente.

**Cundo usar?**

* Este mtodo es particularmente til en teora y simulaciones estadsticas donde la independencia es crucial.
* Es til cuando se trata de poblaciones muy pequeas, ya que permite replicar la poblacin a travs de muestras repetidas.

```python
import numpy as np

# Definir una poblacin
poblacion = np.array([1, 2, 3, 4, 5])

# Muestreo sin reemplazo
muestra_sin_reemplazo = np.random.choice(poblacion, size=3, replace=False)
print("Muestra sin reemplazo:", muestra_sin_reemplazo)

# Muestreo con reemplazo
muestra_con_reemplazo = np.random.choice(poblacion, size=3, replace=True)
print("Muestra con reemplazo:", muestra_con_reemplazo)
```

El muestreo estratificado es un mtodo de muestreo en el que la poblacin se divide en subgrupos o estratos y se toma una muestra de cada estrato.

**cundo usar?**

Este mtodo se utiliza cuando la poblacin es heterognea y se desea que la muestra sea representativa de cada estrato.

**Ventajas:**

* Asegura que cada subgrupo de la poblacin est representado en la muestra.
* Permite comparar subgrupos de la poblacin.
* Aumenta la precisin de la estimacin de parmetros poblacionales.

1. **Dividir la poblacin en estratos**. Los estratos deben ser mutuamente excluyentes y colectivamente exhaustivos, es decir, cada miembro de la poblacin debe pertenecer a un solo estrato y todos los miembros de la poblacin deben pertenecer a al menos un estrato.

2. **Determinar el tamao de la muestra para cada estrato**. El tamao de la muestra para cada estrato debe ser proporcional al tamao del estrato en la poblacin.

3. **Seleccionar una muestra aleatoria de cada estrato**. Seleccionar una muestra aleatoria simple de cada estrato, utilizando mtodos de muestreo aleatorio simple.

## Muestreo por conglomerados

Un conglomerado es un grupo de elementos de la poblacin que comparten caractersticas similares. A diferencia del muestreo estratificado, donde cada estrato debe ser homogneo y los estratos entre s deben ser heterogneos, el muestreo por conglomerados trata de dividir la poblacin en grupos heterogneos que son internamente heterogneos. Cada uno de estos grupos o conglomerados es un pequeo reflejo de la poblacin completa

**Cundo usar?**

Este mtodo se utiliza cuando la poblacin es muy grande y dispersa, y es difcil o costoso seleccionar una muestra aleatoria simple de toda la poblacin.

**Ventajas:**

* Reduce los costos y el tiempo de muestreo.
* Facilita la logstica de la recoleccin de datos.
* Aumenta la eficiencia del muestreo.

**Desventajas:**

* Puede introducir sesgos si los conglomerados no son representativos de la poblacin.
* Puede aumentar la varianza de la estimacin de los parmetros poblacionales.

**Pasos para el muestreo por conglomerados**

1. **Dividir la poblacin en conglomerados**. Los conglomerados deben ser mutuamente excluyentes y colectivamente exhaustivos, es decir, cada miembro de la poblacin debe pertenecer a un solo conglomerado y todos los miembros de la poblacin deben pertenecer a al menos un conglomerado.

2. **Determinar el tamao de la muestra para cada conglomerado**. El tamao de la muestra para cada conglomerado debe ser proporcional al tamao del conglomerado en la poblacin.

3. **Seleccionar una muestra aleatoria de conglomerados**. Seleccionar una muestra aleatoria simple de algunos de los conglomerados, utilizando mtodos de muestreo aleatorio simple.

4. **Seleccionar una muestra aleatoria de elementos dentro de los conglomerados seleccionados**. Seleccionar una muestra aleatoria simple de elementos dentro de los conglomerados seleccionados.

5. **Combinar las muestras de elementos de cada conglomerado**. Combinar las muestras de elementos de cada conglomerado para formar la muestra final.

### Muestreo Sistemtico

El muestreo sistemtico es un mtodo de muestreo en el que los elementos de la poblacin se seleccionan a intervalos regulares. El intervalo de muestreo se calcula dividiendo el tamao de la poblacin por el tamao de la muestra.

**Cundo usar?**

Particularmente til cuando se tiene una lista ordenada de la poblacin y se desea una metodologa de muestreo ms simple que el muestreo aleatorio simple, pero que sigue proporcionando una buena representacin de la poblacin total.

**Ventajas:**

* Es ms eficiente que el muestreo aleatorio simple.
* Es ms fcil de implementar que el muestreo aleatorio simple.

**Desventajas:**

* Puede introducir sesgos si hay un patrn en la poblacin que coincide con el intervalo de muestreo.
* Puede ser menos preciso que el muestreo aleatorio simple si el patrn en la poblacin no es aleatorio.

**Pasos para el muestreo sistemtico**

1. **Definir el Tamao de la Poblacin ($N$) y Ordenar la poblacin** Determinar cuntos elementos hay en total en la poblacin, luego, ordenar la poblacin de acuerdo con un criterio especfico, como el tamao, la edad, la ubicacin, etc.

2. **Determinar el Tamao de la Muestra ($n$):** Decidir cuntos elementos deseas incluir en la muestra.

2. **Calcular el intervalo de muestreo**. Calcular el intervalo de muestreo dividiendo el tamao de la poblacin $(N)$por el tamao de la muestra (n). Se toma el mayor entero al cociente.

$$k = \left\lfloor \frac{N}{n} \right\rfloor$$

3. **Seleccionar un nmero aleatorio entre 1 y $k$**. Seleccionar un nmero aleatorio entre 1 y $k$, este nmero representar el primer elemento de la muestra.

4. **Seleccionar los elementos de la muestra**. Comenzando con el primer elemento seleccionado aleatoriamente se selecciona cada $k$-simo elemento en la lista de la poblacin hasta que hayas seleccionado $n$ elementos.

5. **Combinar los elementos seleccionados para formar la muestra final**. Combinar los elementos seleccionados para formar la muestra final.





# Ejemplo

Supongamos que una empresa quiere realizar una encuesta de satisfaccin entre sus empleados, que son 1000 en total, y decide usar muestreo sistemtico para seleccionar 100 empleados:

* Tamao de la Poblacin $N$: 1000 empleados.
* Tamao de la Muestra $n$: 100 empleados.
* Intervalo de Muestreo $k$:

$$
k = \cfrac{1000}{100} = 10
$$

Punto de Partida Aleatorio: Supongamos que el nmero aleatorio entre 1 y 10 es 7.

Entonces, el muestreo comenzar con el empleado nmero 7 y continuar seleccionando cada dcimo empleado: 7, 17, 27, ..., 997.

```python
# Tamao de la poblacin
N=len(df)

# Tamao de la muestra
n = 5

# Intervalo de muestreo
k=N // n


# Elegir un punto de partida aleatorio entre 0 y k-1
start = np.random.randint(0, k)

# Seleccionar los ndices de la muestra sistemtica
indices = np.arange(start, N, k)

# Seleccionar las filas de la muestra sistemtica en un data frame
muestra_sistematica = df.iloc[indices]

#Imprimir la muestra sistemtica
print("ndices seleccionados:", indices)
print(muestra_sistematica)
```

## Otros muestreos usados en la analtica de datos

### Muestreo de aceptacin

Se genera una muestra de tamao fijo y se acepta o rechaza segn una probabilidad determinada por la relacin entre la funcin de densidad de la distribucin deseada y la funcin de densidad de la distribucin de la muestra.

### Muestreo de reservorio

Se utiliza para seleccionar $n$ muestras de un flujo de tamao desconocido. A medida que llega cada elemento del flujo, se decide si se incluye en la muestra con una probabilidad que depende del nmero de elementos procesados hasta ese momento.

Ideal para flujos de datos (data streams) donde no se conoce el tamao total de los datos de antemano.

### Muestreo de Gibbs

Comn en mtodos de inferencia bayesiana, especialmente til para obtener muestras de distribuciones conjuntas de mltiples variables aleatorias.

Implica muestrear sucesivamente de las distribuciones condicionales de una variable mientras se mantienen fijas las otras, circulando a travs de todas las variables repetidamente.

### Muestreo de bootstrap

Se utiliza para estimar la distribucin de un estadstico de inters mediante el muestreo con reemplazo de la muestra original. Se generan mltiples muestras de la muestra original y se calcula el estadstico de inters en cada muestra.

Es til para estimar la incertidumbre de un estadstico o modelo, especialmente cuando no se conoce la distribucin subyacente de los datos.

* $f(x)$ ser la distribucin de la que queremos obtener una muestra. (Uniforme)
* Constante de Ajuste $c$ tal que  $cg(x)f(x)$ para todo $x$.

```python
import numpy as np
## Ejemplo de muestreo por aceptacin-rechazo

def aceptacion_rechazo(n_samples, lambda_exp):
    """
    Funcin para realizar muestreo por aceptacin-rechazo.

    Args:
    n_samples : int
        Nmero de muestras a generar.
    lambda_exp : float
        Parmetro lambda de la distribucin exponencial deseada.

    Returns:
    np.array
        Array de muestras generadas de la distribucin exponencial.
    """
    samples = []
    c = lambda_exp  # Asumiendo que c debe ser lo suficientemente grande para cubrir la condicin c*g(x) >= f(x).

    while len(samples) < n_samples:
        # Generar una muestra de la distribucin propuesta (uniforme entre 0 y lambda_exp)
        X = np.random.uniform(0, 1 / lambda_exp)

        # Calcular la probabilidad de aceptacin
        prob_aceptacion = lambda_exp * np.exp(-lambda_exp * X)

        # Decidir si aceptar o rechazar la muestra
        U = np.random.uniform(0, 1)
        if U < prob_aceptacion:
            samples.append(X)

    return np.array(samples)
```

```python
# Generar 1000 muestras de una distribucin exponencial con lambda = 0.5
lambda_param = 0.5
generated_samples = aceptacion_rechazo(1000, lambda_param)

print(generated_samples[:10])  # Mostrar las primeras 10 muestras
```
